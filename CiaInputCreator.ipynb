```python
import json
import re
```

```python
cia = []
with open("../factbook.json") as infile:
    for line in infile:
        cia_fact = json.loads(line)
        cia.append(cia_fact)
```

```python
catArr = [] #stores categories in order
catArr.append("Is a republic")
catArr.append("Is a democracy")
catArr.append("Is a parliamentary government")
catArr.append("Is a territory of another country")
catArr.append("Is another type of government")
catArr.append("GDP in millions")
catArr.append("GDP in billions")
catArr.append("GDP in trillions")
catArr.append("Has textiles industry")
catArr.append("Has oil industry")
catArr.append("Has clothing industry")
catArr.append("Has food processing industry")
catArr.append("Has chemical industry")
catArr.append("Has mining industry")
catArr.append("Has software industry")
catArr.append("Has low unemployment rate")
catArr.append("Has high unemployment rate")
catArr.append("Has budget surplus")
catArr.append("Has budget deficit")
```

```python
datArr = [] #stores first round of data
countryArr = [] #stores country names in order
currCount = 0
for country in cia:
    tempArr = []
    try:
        countryArr.append(country['countryname'])
    except:
        countryArr.append("none")
     
    try:
        gov = country['sections']['Government'][1]['category_data']
        if "republic" in gov:
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(0)
        elif "democracy" in gov:
            tempArr.append(0)
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
        elif "parliamentary" in gov:
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(1)
            tempArr.append(0)
        else:
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(1)
    except:
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        
    try:
        gdp = country['sections']['Economy'][2]['category_data'].split('(')[0]
        gdp = gdp.split(' ')[1]
        if gdp == "million":
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
        elif gdp == "billion":
            tempArr.append(0)
            tempArr.append(1)
            tempArr.append(0)
        elif gdp == "trillion":
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(1)
        else:
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(0)
    except:
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        
    try:
        industries = country['sections']['Economy'][9]['category_data'].split(',')
        for a in range(len(industries)):
            industries[a] = industries[a].strip()
        if "textiles" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "oil" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "clothing and footwear" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "food processing" in industries or "food and beverages" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "chemicals" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "mining" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
        if "software" in industries:
            tempArr.append(1)
        else:
            tempArr.append(0)
    except:
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)
        tempArr.append(0)

    try:
        unemp = country['sections']['Economy'][13]['category_data'].split('%')[0]
        if float(unemp) < 20:
            tempArr.append(1)
            tempArr.append(0)
        else:
            tempArr.append(0)
            tempArr.append(1)
    except:
        tempArr.append(0)
        tempArr.append(0)
        
    try:
        bud = country['sections']['Economy'][19]['category_data'].split('%')[0]
        if float(bud) > 0:
            tempArr.append(1)
            tempArr.append(0)
        else:
            tempArr.append(0)
            tempArr.append(1)
    except:
        tempArr.append(0)
        tempArr.append(0)
        
    datArr.append(tempArr)
    currCount = currCount + 1
    
```

```python
#takes countryArr and turns it into a 1D array where array[1] = "0 3 4 7 10 11..."
count = 0
finArr = []
for a in datArr:
    tempStr = ""
    for b in range(len(a)):
        if a[b] == 1:
                tempStr = tempStr + str(b) + ' '
    tempStr = tempStr[:-1]
    finArr.append(tempStr)
```

```python
f = open('data_test.dat', 'w')
for a in finArr:
    f.write(a + '\n')
```
