```python
import json
import re
```

```python
cia = []
with open("../factbook.json") as infile:
    for line in infile:
        cia_fact = json.loads(line)
        cia.append(cia_fact)
```

```python
catArr = [] #stores categories in order
catArr.append("Is a republic")
catArr.append("Is a democracy")
catArr.append("Is a parliamentary government")
catArr.append("Is a territory of another country")
catArr.append("Is another type of government")
catArr.append("GDP in millions")
catArr.append("GDP in billions")
catArr.append("GDP in trillions")
```

```python
datArr = [] #stores first round of data
countryArr = [] #stores country names in order
currCount = 0
for country in cia:
    tempArr = []
    try:
        countryArr.append(country['countryname'])
    except:
        countryArr.append("none")
     
    try:
        gov = country['sections']['Government'][1]['category_data']
        if "republic" in gov:
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(0)
        elif "democracy" in gov:
            tempArr.append(0)
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
        elif "parliamentary" in gov:
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(1)
            tempArr.append(0)
        else:
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(0)
            tempArr.append(1)
    except:
        print(" ") 
        
    try:
        gdp = country['sections']['Economy'][2]['category_data'].split('(')[0]
        gdp = gdp.split(' ')[1]
        if gdp == "million":
            tempArr.append(1)
            tempArr.append(0)
            tempArr.append(0)
        elif gdp == "billion":
            datArr[currCount][5] = 0
            datArr[currCount][6] = 1
            datArr[currCount][7] = 0
        elif gdp == "trillion":
            datArr[currCount][5] = 0
            datArr[currCount][6] = 0
            datArr[currCount][7] = 1
    except:
        print(" ")
        
    currCount = currCount + 1
    
```
